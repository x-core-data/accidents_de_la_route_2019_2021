### Chargement des librarires nécessaires à l'exploitation des données :

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
```

### Chargement des bases de données liées aux véhicules avec la méthode "read_csv" de pandas pour les années 2017 à 2021 :


```python
df_vehicules_2021 = pd.read_csv("data/vehicules-2021.csv", sep=';')
df_vehicules_2020 = pd.read_csv("data/vehicules-2020.csv", sep=';')
df_vehicules_2019 = pd.read_csv("data/vehicules-2019.csv", sep=';')
```

### Fusion et noettoyage des dataframes :

```python
df = [df_vehicules_2019, df_vehicules_2020, df_vehicules_2021]
vehicules_2019_2021 = pd.concat(df)
```

```python
sns.heatmap(vehicules_2019_2021.isna());
```

```python
#remplacement des -1 (non renseigné) par des NaN(Not a Number)
vehicules_2019_2021 = vehicules_2019_2021.replace(-1, np.nan)
```

```python
sns.heatmap(vehicules_2019_2021.isna());
```

```python
#remplacement des 0 (sans objet) par des NaN(Not a Number)
vehicules_2019_2021 = vehicules_2019_2021.replace(0, np.nan)
```

```python
sns.heatmap(vehicules_2019_2021.isna());
```

```python
vehicules_2019_2021.to_csv('merged_tables/vehicules_2019_2021.csv', index=False)
```
