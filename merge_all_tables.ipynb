### Chargement de l'ensemble des bases de données :

```python
caracteristiques_2019_2021 = pd.read_csv("merged_tables/caracteristiques_2019_2021.csv", sep=',')
lieux_2019_2021 = pd.read_csv("merged_tables/lieux_2019_2021.csv", sep=',')
usagers_2019_2021 = pd.read_csv("merged_tables/usagers_2019_2021.csv", sep=',')
vehicules_2019_2021 = pd.read_csv("merged_tables/vehicules_2019_2021.csv", sep=',')
```

```python
user_carac = caracteristiques_2019_2021.merge(usagers_2019_2021, on='Num_Acc')
```

```python
usagers_2019_2021
```

```python
user_carac.to_csv('usercarac.csv', index=False)
```

```python
user_carac
```

```python
user_carac_lieux = user_carac.merge(lieux_2019_2021, on='Num_Acc')
```

```python
user_carac_lieux.to_csv('userrrr.csv', index=False)
```

```python
final_df = user_carac_lieux.merge(vehicules_2019_2021, on='id_vehicule')
```

```python
final_df.to_csv('finaldf.csv', index=False)
```
